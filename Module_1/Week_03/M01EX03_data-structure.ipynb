```python
# Test softmax function in torch
data = torch.Tensor([1, 2, 3])
softmax = nn.Softmax(dim=0)
print(softmax(data))
```

```python
# create my softmax function
class MySoftmax(nn.Module):
    def __init__(self):
        super().__init__()
    
    def forward(self, x):
        x_exp = torch.exp(x)
        partition = torch.sum(x_exp)
        return x_exp / partition

my_softmax = MySoftmax()
print(my_softmax(data))
```

```python
# create softmax stable function
class MySoftmaxStable(nn.Module):
    def __init__(self):
        super().__init__()
    
    def forward(self, x):
        x_exp = torch.exp(x - torch.max(x))
        partition = torch.sum(x_exp)
        return x_exp / partition

my_softmax_stable = MySoftmaxStable()
print(my_softmax_stable(data))
```

```python
from abc import ABC, abstractmethod

# Create a class Person with an abstract method describe
class Person:
    def __init__(self, name, yob):
        self.name = name
        self.yob = yob
    
    def get_yob(self):
        return self.yob
    
    @abstractmethod
    def describe(self):
        pass

class Student(Person):
    def __init__(self, name:str, yob:int, grade:str):
        super().__init__(name=name, yob=yob)
        self.grade = grade
    
    def describe(self):
        print(f"Student {self.name} was born in {self.yob} and is in grade {self.grade}")

class Teacher(Person):
    def __init__(self, name:str, yob:int, subject:str):
        super().__init__(name=name, yob=yob)
        self.subject = subject
    
    def describe(self):
        print(f"Teacher {self.name} was born in {self.yob} and teaches {self.subject}")

class Doctor(Person):
    def __init__(self, name:str, yob:int, specialist:str):
        super().__init__(name=name, yob=yob)
        self.specialist = specialist
    
    def describe(self):
        print(f"Doctor {self.name} was born in {self.yob} and specializes in {self.specialist}")
```

```python
# Test the classes
student1 = Student(name="studentA", yob=2010, grade="7")
student1.describe()

teacher1 = Teacher(name="teacherA", yob=1969, subject="Math")
teacher1.describe()

doctor1 = Doctor(name="doctorA", yob=1945, specialist="Endocrinologists")
doctor1.describe()
```

```python
# Create a class Ward
class Ward:
    def __init__(self, name:str):
        self.name = name
        self.__list_people = list()
    
    def add_person(self, person:Person):
        self.__list_people.append(person)

    def describe(self):
        print(f"List of people in the ward {self.name}")
        for person in self.__list_people:
            person.describe()
    
    def count_doctors(self):
        count = 0
        for person in self.__list_people:
            if isinstance(person, Doctor):
                count += 1
        return count
    
    def sort_age(self):
        self.__list_people.sort(key=lambda x: x.get_yob(), reverse=True)
    
    def compoute_avg_age(self):
        counter = 0
        total_year = 0
        for p in self.__list_people:
            if isinstance(p, Teacher): 
                counter += 1
                total_year += p.get_yob()
        return total_year/counter
```

```python
# Test the class Ward
teacher2 = Teacher(name="teacherB", yob=1995, subject="History")
doctor2 = Doctor(name="doctorB", yob=1975, specialist="Cardiologists")
ward1 = Ward(name="Ward1")
ward1.add_person(student1)
ward1.add_person(teacher1)
ward1.add_person(teacher2)
ward1.add_person(doctor1)
ward1.add_person(doctor2)
ward1.describe()
```

```python
# Test the count_doctors method
print(f"\nNumber of doctors: {ward1.count_doctors()}")
```

```python
# Test the sort_age method
print("\nAfter sorting Age of Ward1 people")
ward1.sort_age()
ward1.describe()
```

```python
# Test the compute_average method
print(f"\nAverage year of birth (teachers): {ward1.compoute_avg_age()}")
```

```python
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = []

    def is_empty(self):
        return len(self.stack) == 0

    def is_full(self):
        return len(self.stack) == self.capacity

    def pop(self):
        if not self.is_empty():
            return self.stack.pop()
        else:
            return "Stack is empty"

    def push(self, value):
        if not self.is_full():
            self.stack.append(value)
        else:
            return "Stack is full"

    def top(self):
        if not self.is_empty():
            return self.stack[-1]
        else:
            return "Stack is empty"
```

```python
#  Test the Stack class

stack1 = Stack(capacity=5)

stack1.push(1)

stack1.push(2)

print(stack1.is_full())

print(stack1.top())

print(stack1.pop())

print(stack1.top())

print(stack1.pop())

print(stack1.is_empty())
```
